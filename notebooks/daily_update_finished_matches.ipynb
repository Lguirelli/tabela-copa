# Atualização diária de jogos finalizados — sem API key e sem proxy

Este notebook executa o script `scripts/daily_extract_finished_matches.py`. Ele coleta apenas dados diretos de fontes públicas sem chave e não estima campos ausentes.

```python
from pathlib import Path
import subprocess, sys

repo_root = Path.cwd()
while not (repo_root / "scripts" / "daily_extract_finished_matches.py").exists() and repo_root.parent != repo_root:
    repo_root = repo_root.parent
print("Repo root:", repo_root)
```

```python
cmd = [sys.executable, str(repo_root / "scripts" / "daily_extract_finished_matches.py"), "--repo-root", str(repo_root), "--days-back", "3", "--days-forward", "0"]
print("Rodando:", " ".join(cmd))
subprocess.run(cmd, check=True)
```

Saídas geradas em `data/daily_updates/` e resultados atualizados em `data/resultados_reais.csv`, `data/resultados.csv` e `data/database/matches.csv`.
